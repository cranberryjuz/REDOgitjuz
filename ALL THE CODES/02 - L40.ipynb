## Data Types in Python

#### Float & Integer

```python
a = 13.14
b = 4896
print("a =",a,"is",type(a))
print("b =",b,"is",type(b))
```

#### String

```python
sentence_str = "Handsome Koh is a really handsome guy."
sentence_lst = sentence_str.split()
print(sentence_str)
print(sentence_lst)
```

#### Boolean

```python
num1 = 4896
num2 = 0
num3 = -1

print(bool(num1),bool(num2),bool(num3))
```

```python
weekdays_tup = ('Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday')
weekdays_tup
```

```python
weekdays_lst = list(weekdays_tup)
weekdays_lst
```

```python
str1 = "Handsome"
str2 = 'Koh'
str3 = 13.14
str4 = 4896
str5 = True

record_lst = list()
record_lst.append(str1)
record_lst.append(str2)
record_lst.append(str3)
record_lst.append(str4)
record_lst.append(str5)
record_lst
```

```python
record_dict = dict()
record_dict['first'] = str1
record_dict['last'] = str2
record_dict['rating'] = str3
record_dict['salary'] = str4
record_dict['executive'] = str5
record_dict
```

## Dealing with File

#### Import plain text file

open file in Python has 3 modes:
1. r - read only
2. w - write / overwrite
3. a - append / write at the bottom

```python
sentence_file = open('Sentence.txt','r')
sentence_file
```

```python
sentence_str = sentence_file.read()
sentence_str
```

```python
sentence_lst = sentence_str.split()
sentence_lst
```
